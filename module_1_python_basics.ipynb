```python
## 1. Python Data types
Python distinguishes between several basic data types:

Integer (int)
Float (float)
String (str)
Boolean (bool)

Understanding data types is essential for analysis,
as they define how data can be stored, manipulated and converted.
```

```python
#Integer
age = 42

#Float
height = 1.65

#Boolean
is_student = True

#String
name = "Lourdes"

type(age), type(height), type(is_student), type(name)

--> What this block demonstrates <--

·You know how to define variables
·You know how to identify data types
·You use type() to validate them
```

```python
## 2. Type Conversion

Python allows converting one data type into another.
This is useful, for example, when numeric data is read as text and needs to be converted.
```

```python
# Integer to float
age_float = float(age)

# Float to integer
height_int = int(height)

# Numeric to string
age_str = str(age)

# Numeric to boolean
zero_value = 0
non_zero_value = 5

bool(zero_value), bool(non_zero_value)
```

```python
## 3. Expressions and Operators

Expressions are combinations of values and operations that produce a result.
Python follows the order of operations (BODMAS)
```

```python
#Operations

x = 9
y = 4

addition = x + y
subtraction = x - y
multiplication = x * y
division = x / y
floor_division = x // y
modulus = x % y 

addition, subtraction, multiplication, division, floor_division, modulus
```
