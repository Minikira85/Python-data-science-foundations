```python


## 3. Expressions and Operatos

Expressions adn combinations of values and operations that produce a result.
Python follows the order of operations (BODMAS)          
```

```python
#Operations

x = 9
y = 4

addition = x + y
substration = x - y
multiplication = x * y
division = x / y
floor_division = x // y
module = x % y 

addition, substration, multiplication, division, floor_division, module
```

```python


## 4. Variables Assignment and Reassignment

Variables can be updated during program execution.
When a variable is reassigned, its previous value is replaced by the new one.
```

```python


# Initial assignment
score = 10

# Reassignment: updating the value
score = score + 5

# Display the updated value
score

score = 10
print("Initial score:", score)

score = score + 5
print("Updated score:", score)
```

```python


## 5. String Operations

Strings are sequences of characters.
They support indexing, slicing, concatenation, and formatting.
```

```python


text = "Behavioral Data Science"
# Indexing
first_letter = text[0]
last_letter = text[-1]

# Slicing
word = text[0:10]

first_letter, last_letter, word
```

```python

## 6. String Methods

Python provides built-in methods to manipulate text data, such as changing case,
replacing text, or removing whitespace.
```

```python


message = "  Hello World  "

message_upper = message.upper()
message_lower = message.lower()
message_stripped = message.strip()
message_replaced = message.replace("World", "Python")

message_upper, message_lower, message_stripped, message_replaced
```
